# Primary Features' extraction from click-stream dump

In this code we aim to extract features for later prediction of improvement in assigment based on the pickle file that has all the data of student's actions between submissions.

```python
import os, time, pickle, csv
from tabulate import tabulate
from termcolor import colored
from CustomFeatures import CalculateFeatures #Here comes the costume
import CreateTable_Utilities as ctu
import pandas as pd
import numpy as np
%matplotlib inline
```

### 1. Loading given data and custum features

Get the data into python

```python
# Get input dataset filepath
FilePath1 = 'dataset.pickle'
FilePath2 = 'dataset_test.pickle'


# Load dataset into python dictionary
with open(FilePath1, 'rb') as handle:
    DataSet1 = pickle.load(handle)
with open(FilePath2, 'rb') as handle:
    DataSet2 = pickle.load(handle)
```

Extract features from click-stream and save into intermediate CSV

```python
# Add custom features to data set
DataSet12 = ctu.AppendFeatures(DataSet1)
DataSet22 = ctu.AppendFeatures(DataSet2)


# Save data set into CSV file
ctu.ExportAsCSV(DataSet12,"train.csv")
ctu.ExportAsCSV(DataSet22,"test.csv")
```

### 2. Cleaning

Import into Panda's DataFrame and perform basic cleaning and filtering

```python
# Import
df_tr = pd.read_csv('train.csv', sep=',')
df_te = pd.read_csv('test.csv', sep=',')
```

```python
# Filter by number of submissions
df_tr_ms = df_tr.query('SubmissionNumber>0')
df_te_ms = df_te.query('SubmissionNumber>0')
```

```python
# Replace nan by 0
df_tr_clean = df_tr_ms.fillna(value=0)
df_te_clean = df_te_ms.fillna(value=0)
```

### 3. Transformation

Some transformations can be useful

We also want to create a meaningful index that conveys which student is acting on which problem

```python
# Create composite index 1 (for first prob.)
df_tr_clean['uniqRowID_1'] = df_tr_ms.UserID + '_' + df_tr_ms.ProblemID.astype('str') + '_' + df_tr_ms.SubmissionNumber.astype('str')
df_te_clean['uniqRowID_1'] = df_te_ms.UserID + '_' + df_te_ms.ProblemID.astype('str') + '_' + df_te_ms.SubmissionNumber.astype('str')
```

```python
# Create composite index 2 (will be helpful when grouping for secondary features)
df_tr_clean['uniqRowID_2'] = df_tr_ms.UserID + '_' + df_tr_ms.ProblemID.astype('str')
df_te_clean['uniqRowID_2'] = df_te_ms.UserID + '_' + df_te_ms.ProblemID.astype('str')
```

```python
#Problem initial diff.
df = df_tr.ix[:,["UserID","ProblemID","Grade","SubmissionNumber","GradeDiff"]]
df_gradeini = df.drop_duplicates(subset=["UserID","ProblemID"],keep='first')
df_gradefin = df.drop_duplicates(subset=["UserID","ProblemID"],keep='last')

#Possible features by prob
diff_ini_prob = df_gradeini.groupby("ProblemID")[['Grade']].mean()
diff_ini_prob.rename(columns={"Grade":"Diff_Prob_Ini"},inplace=True)
diff_fin_prob = df_gradefin.groupby("ProblemID")[['Grade']].mean()
diff_fin_prob.rename(columns={"Grade":"Diff_Prob_Fin"},inplace=True)
av_improvement_sub_prob = df.query("SubmissionNumber>0").groupby("ProblemID")[['GradeDiff']].mean()
av_improvement_sub_prob.rename(columns={"GradeDiff":"Exp_Prob_GradeDiff"},inplace=True)
features_by_prob = pd.concat([diff_ini_prob, diff_fin_prob, av_improvement_sub_prob], axis=1).reset_index()
features_by_prob.head()
```

```python
#Nice finding: No one resubmmits problem 16
df.query("SubmissionNumber>0 & ProblemID==16").head()
```

```python
#Merge with past data
df_tr_extra = pd.merge(df_tr_clean, features_by_prob, how='left', on=['ProblemID'])
df_te_extra = pd.merge(df_te_clean, features_by_prob, how='left', on=['ProblemID'])
```

Now some extra additions pre-simplifications

```python
df_tr_extra['logTimeSinceLast'] = np.log(df_tr_extra.TimeSinceLast)
df_te_extra['logTimeSinceLast'] = np.log(df_te_extra.TimeSinceLast)

# Trying binarization

# df_tr_extra['Video'] = (df_tr_extra.NVideoEvents >0) *1.0
# df_tr_extra['Forum'] = (df_tr_extra.NForumEvents >0) *1.0

# df_te_extra['Video'] = (df_te_extra.NVideoEvents >0) *1.0
# df_te_extra['Forum'] = (df_te_extra.NForumEvents >0) *1.0
```

```python
df_tr_final = df_tr_extra.drop(['UserID','TimeStamp','Grade'], 1)
df_te_final = df_te_extra.drop(['UserID','TimeStamp','Grade'], 1)
df_tr_final.head()
```
